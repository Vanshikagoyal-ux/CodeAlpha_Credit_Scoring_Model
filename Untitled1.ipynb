```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, roc_auc_score


df = pd.read_csv('application_record.csv')

df['TARGET'] = ((df['DAYS_EMPLOYED'] > 0) | (df['CNT_CHILDREN'] > 2)).astype(int)


X = df[['AMT_INCOME_TOTAL', 'DAYS_BIRTH', 'CNT_CHILDREN']]
y = df['TARGET']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


print("...")
model = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42)
model.fit(X_train, y_train)


predictions = model.predict(X_test)
prob_predictions = model.predict_proba(X_test)[:, 1]

print("\n---  ---")
print(classification_report(y_test, predictions))
print(f"ROC-AUC Score: {roc_auc_score(y_test, prob_predictions):.2f}")
```
